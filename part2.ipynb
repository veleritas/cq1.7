## Function for querying and grabbing associated phenotypes

```python
def query(gene_id):
    """Query Monarch to determine the phenotypes associated with a
    NCBI gene id."""
    
    url = "https://api.monarchinitiative.org/api/bioentity/gene/{}/phenotypes/".format(gene_id)
    
    res = requests.get(url)
    return res.json()["objects"]
```

Assuming that the "objects" reponse object contains the phenotypes.

## Grab phenotypes for our genes

```python
phenos = {}
for gene_id in tqdm(genes):
    res = query(gene_id)
    phenos[gene_id] = res
```
